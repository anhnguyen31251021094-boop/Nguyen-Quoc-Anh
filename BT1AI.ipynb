```python
import folium
from geopy.geocoders import Nominatim
import osmnx as ox
import networkx as nx
# Locations
address_A = "Ben Thanh Market, Ho Chi Minh City, Vietnam"
address_B = "Tan Son Nhat Airport, Ho Chi Minh City, Vietnam"
# Get coordinates
geolocator = Nominatim(user_agent="routing-example", timeout=10) # Added timeout
location_A = geolocator.geocode(address_A)
location_B = geolocator.geocode(address_B)
# Check if locations were found before proceeding
latlng_A = (location_A.latitude, location_A.longitude)
latlng_B = (location_B.latitude, location_B.longitude)
# Download road network (graph) from OpenStreetMap
# Note: we take the road network in the area between A and B, radius about 3km
midpoint = ((latlng_A[0] + latlng_B[0]) / 2, (latlng_A[1] + latlng_B[1]) / 2)
G = ox.graph_from_point(midpoint, dist=3000, network_type='drive')
# Find the closest node to A and B
orig_node = ox.distance.nearest_nodes(G, latlng_A[1], latlng_A[0])
dest_node = ox.distance.nearest_nodes(G, latlng_B[1], latlng_B[0])
# Find the shortest path by length
route = nx.shortest_path(G, orig_node, dest_node, weight='length')
# Route coordinates
route_coords = [(G.nodes[n]['y'], G.nodes[n]['x']) for n in route]
# Create a map
m = folium.Map(location=midpoint, zoom_start=14)
# Add markers A and B
folium.Marker(latlng_A, popup="Điểm A", icon=folium.Icon(color='red')).add_to(m)
folium.Marker(latlng_B, popup="Điểm B", icon=folium.Icon(color='blue')).add_to(m)
# Draw the route according to reality
folium.PolyLine(locations=route_coords, color='green', weight=5, opacity=0.8).add_to(m)
# Show map
display(m)
```

```python
import folium
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import time

geolocator = Nominatim(user_agent="my_map_app")


addresses = [
    "Ben Thanh Market, Ho Chi Minh City",
    "Landmark 81, Ho Chi Minh City",
    "Tan Son Nhat Airport, Ho Chi Minh City",
    "Saigon Zoo, Ho Chi Minh City",
    "Independence Palace, Ho Chi Minh City",
    "Notre Dame Cathedral, Ho Chi Minh City",
    "Nguyen Hue Walking Street, Ho Chi Minh City",
    "University of Technology HCM",
    "Pham Ngu Lao Street, Ho Chi Minh City",
    "District 7, Ho Chi Minh City"
]


coords = []
names = []

for addr in addresses:
    location = geolocator.geocode(addr)
    if location:
        coords.append((location.latitude, location.longitude))
        names.append(addr)
    else:
        print(f"Không tìm thấy: {addr}")
    time.sleep(1)


center = coords[0]


m = folium.Map(location=center, zoom_start=13)


for i, (c, name) in enumerate(zip(coords, names)):
    distance = geodesic(center, c).km


    folium.Marker(
        location=c,
        popup=f"{name}\nKhoảng cách: {round(distance,2)} km",
        tooltip=f"Điểm {i+1}"
    ).add_to(m)


    folium.PolyLine(
        locations=[center, c],
        color="blue"
    ).add_to(m)
m.save("map.html")


m
```

```python

import osmnx as ox
import matplotlib.pyplot as plt


place = "District 1, Ho Chi Minh City, Vietnam"


G = ox.graph_from_place(place, network_type="drive")


ox.plot_graph(G)


num_nodes = len(G.nodes)
num_edges = len(G.edges)

lengths = [data["length"] for u, v, data in G.edges(data=True)]
avg_length = sum(lengths) / len(lengths)

print("Số nút giao:", num_nodes)
print("Số đoạn đường:", num_edges)
print("Chiều dài trung bình (m):", avg_length)



area = ox.geocode_to_gdf(place).area.values[0]
density = num_edges / area

print("Mật độ mạng:", density)
```

```python
!pip install --upgrade osmnx networkx
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt


place = "District 1, Ho Chi Minh City, Vietnam"
G = ox.graph_from_place(place, network_type="drive")


Q1 = (10.7769, 106.7009)
Q3 = (10.7820, 106.6950)


Q1_node = ox.distance.nearest_nodes(G, Q1[1], Q1[0])
dest_node = ox.distance.nearest_nodes(G, Q3[1], Q3[0])

route_dijkstra = nx.shortest_path(G, Q1_node, dest_node, weight="length")

def heuristic(u, v):
    x1, y1 = G.nodes[u]['x'], G.nodes[u]['y']
    x2, y2 = G.nodes[v]['x'], G.nodes[v]['y']
    return ((x1 - x2)**2 + (y1 - y2)**2)**0.5

route_astar = nx.astar_path(G, Q1_node, dest_node, heuristic=heuristic, weight="length")

fig, ax = ox.plot_graph_routes(
    G,
    routes=[route_dijkstra, route_astar],
    route_colors=["red", "blue"],
    route_linewidth=4,
    node_size=0
)

# Calculate length for Dijkstra route
len_dijkstra = 0
for i in range(len(route_dijkstra) - 1):
    u = route_dijkstra[i]
    v = route_dijkstra[i+1]
    edge_attr = G.get_edge_data(u, v)
    if edge_attr and 0 in edge_attr:
        len_dijkstra += edge_attr[0]['length']

# Calculate length for A* route
len_astar = 0
for i in range(len(route_astar) - 1):
    u = route_astar[i]
    v = route_astar[i+1]
    edge_attr = G.get_edge_data(u, v)
    if edge_attr and 0 in edge_attr:
        len_astar += edge_attr[0]['length']

print("Dijkstra length (m):", len_dijkstra)
print("A* length (m):", len_astar)
```
